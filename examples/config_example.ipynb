# Config Object

This is an example of how to use the `mt_metadata` style Configuration object.


```python
import json
from aurora.config.config_creator import ConfigCreator
```

```python
c = ConfigCreator()
p = c.create_run_processing_object()
```

## Add in specific values

```python
p.id = "test1-fortran"
p.stations.local.mth5_path = "/home/kkappler/software/irismt/aurora/tests/synthetic/data/test1.h5"
p.stations.local.id = "test1"

for decimation in p.decimations:
    decimation.estimator.engine = "RME"
    decimation.window.type = "hamming"
    decimation.regression.max_redescending_iterations = 2
```

```python
p
```

```python
with open("example_config.json", "w") as fid:
    json.dump(p.to_dict(), fid)
```
